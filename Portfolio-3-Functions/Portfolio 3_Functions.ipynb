## Custom Pizza Ordering System
<div style="text-align: justify;">
This portfolio features a Pizza Ordering and Customization System, an automated program designed to build and calculate a customer’s pizza order based on selected preferences. The project focuses on the use of Functions to organize the program into reusable parts, following a clear sequence of steps to process inputs and generate results. Key concepts applied include defining functions such as crust selection, topping additions, and discount application, the use of parameters to handle varying user inputs, and fruitful functions to return computed values for total cost calculation. It also incorporates conditionals and numeric expressions to determine pricing, along with void functions to display the final receipt, and a try/except structure to handle invalid or non-numeric user input, ensuring the system remains reliable and user-friendly.

#### Constants
Python constants store fixed values used throughout the program. Clear names improve readability.

```python
SHOP_NAME = "In Crust We Trust"
TAX_RATE = 0.12  # 12% Sales Tax
```

#### Welcome Display
A void function displays output without returning a value.

```python
def display_welcome():
    print("*****************************")
    print("  Welcome to", SHOP_NAME)
    print("*****************************")
```

#### Crust Selection
Uses parameters and conditionals to return the selected crust price.

```python
def select_crust(choice):
    if choice == '1':
        return 100.0
    elif choice == '2':
        return 150.0
    elif choice == '3':
        return 200.0
    else:
        return 0.0
```

#### Toppings Calculation
Calculates total cost based on the number of toppings.

```python
def calculate_toppings(quantity):
    cost = float(quantity) * 30.0
    return cost
```

#### Size Multiplier
Adjusts the price depending on the selected pizza size.

```python
def apply_size_multiplier(base_cost, size_choice):
    if size_choice == 'S':
        return base_cost * 1.0
    elif size_choice == 'M':
        return base_cost * 1.5
    elif size_choice == 'L':
        return base_cost * 2.0
    else:
        return base_cost
```

#### Discount
Applies a discount if the subtotal exceeds a certain amount.

```python
def check_discount(current_total):
    if current_total > 400.0:
        print(">> Promo: 50.0 discount applied!")
        return current_total - 50.0
    return current_total
```

#### Tax Computation
Computes the tax using a fixed rate.

```python
def compute_tax(amount):
    return amount * TAX_RATE
```

#### Receipt Printing
Displays the final order summary using multiple arguments.

```python
def print_receipt(sub, tx, total):
    print("\n--- FINAL RECEIPT ---")
    print("Subtotal: ", sub)
    print("Tax:      ", tx)
    print("---------------------")
    print("TOTAL:    ", total)
    print("---------------------")
```

#### Input Function
Collects user input and organizes the order process.

```python
def run_order_system():
    display_welcome()

    print("Crust Options: 1: Thin(100), 2: Thick(150), 3: Stuffed(200)")
    c_input = input("Choose Crust (1-3): ")
    crust_price = select_crust(c_input)

    if crust_price == 0.0:
        print("Invalid crust selected. Ending program.")
        return

    size = input("Enter Size (S, M, L): ")
    top_qty = input("How many extra toppings? ")

    return crust_price, size, top_qty
```

#### Order Processing
Performs all calculations using functions and expressions.

```python
def process_order(crust_price, size, top_qty):
    topping_total = calculate_toppings(top_qty)

    initial_subtotal = apply_size_multiplier(crust_price, size) + topping_total
    discounted_subtotal = check_discount(initial_subtotal)

    tax_amount = compute_tax(discounted_subtotal)
    final_bill = discounted_subtotal + tax_amount

    return discounted_subtotal, tax_amount, final_bill
```

#### Execution and Invoice
Runs the program and handles invalid input using try/except; displays the final summary and total cost.

```python
try:
    result = run_order_system()

    if result:
        crust_price, size, top_qty = result
        sub, tax, total = process_order(crust_price, size, top_qty)

        print_receipt(sub, tax, total)

except:
    print("Error: Please enter a numeric value for toppings.")
    
print("End of Session.")
```
