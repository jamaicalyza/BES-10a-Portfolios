## Sisig Recipe Scaler
<div style="text-align: justify;">
This portfolio presents a sisig recipe scaling program designed to adjust ingredient quantities for a varying number of servings, helping users accurately prepare meals for different group sizes. It demonstrates the use of iterations alongside concepts from expressions, conditionals, and functions by defining a scaling function, using loops to repeatedly accept and process ingredient inputs, and applying filtering logic to display warnings when quantities become excessively large. The program also includes a try/except structure to ensure valid numeric input for serving sizes, creating a practical and user-friendly system that reflects logical program flow and real-world application.

#### Function
This function multiplies ingredient amounts based on number of guests.

```python
def scale_quantity(original_amount, guest_count):
    return original_amount * guest_count
```

#### Variables
Initializes counters, storage list, and warning flag.

```python
total_ingredients = 0
ingredient_list = []
found_large_item = False
```

#### Title
Displays program header.

```python
print('--- Sisig Recipe Scaler ---')
```

#### Guest
Gets total number of guests and ensures valid input.

```python
try:
    servings = input('Enter number of guests: ')
    servings_int = int(servings)
except:
    print('Error: Please enter a valid number.')
    quit()
```

#### Ingredient
Collects ingredient name, unit, and base amount until user types "done".

```python
while True:
    item_name = input('Enter ingredient name (or type "done"): ')

    if item_name.strip().lower() == 'done':
        break

    unit = input('Enter unit (g, mL, tbsp): ')
    raw_amount = input(f'Enter base amount in {unit}: ')

    try:
        amount_float = float(raw_amount)
    except:
        print('Invalid amount. Skipping ingredient...')
        continue

    scaled_amount = scale_quantity(amount_float, servings_int)

    ingredient_list.append((item_name, scaled_amount, unit))

    if scaled_amount > 1000:
        found_large_item = True

    total_ingredients += 1
```

#### Summary
Prints all scaled ingredients and final program summary.

```python
for item_name, scaled_amount, unit in ingredient_list:
    print('Scaled', item_name, ':', scaled_amount, unit)

print('--- Scaling Complete ---')
print('Total ingredients processed:', total_ingredients)

if found_large_item:
    print('Note: Some quantities require heavy-duty equipment.')
else:
    print('All quantities are standard sizes.')
```
