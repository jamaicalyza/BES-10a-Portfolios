## Shipping Logistics System
This portfolio features a Shipping Logistics System which is an automated tool designed to determine total delivery costs by evaluating a customer's specific shipment details. The project uses Conditionals along with basic concepts from the Intro and Expressions lessons, and follows a clear order of steps to calculate costs using numeric expressions. Key topics applied include one-way decisions to check data, two-way decisions using if and else, multi-way branching using elif to group package weights, and nested decisions for extra distance fees. It also uses a try/except structure to handle non-numeric user input.

#### Global Constants (Fixed Values)
Python constants are fixed values. Mnemonic names make the code easier to read.

```python
BASE_FEE = 7.00     # Standard handling fee
PRIORITY_FEE = 15.00 # Extra fee for express service
VAT_RATE = 0.12      # 12% Value Added Tax
```

#### Weight Input & Conversion
Reads user input and converts it to a numeric value. A try/except block handles invalid input. 

```python
raw_weight = input('Enter package weight (kg): ')

try:
    weight = float(raw_weight) # Type conversion to float
    print('Weight validated:', weight)
except:
    print('Error: Invalid numeric input for weight.')
    weight = -1 # Sentinel value to indicate failure
```

#### Distance Input & Conversion
Reads distance input and converts it to a numeric value. A try/except block handles invalid input.

```python
raw_dist = input('Enter shipping distance (km): ')

try:
    distance = float(raw_dist) # Type conversion to float
    print('Distance validated:', distance)
except:
    print('Error: Invalid numeric input for distance.')
    distance = -1
```

#### Service Type
Reads the service type as text input. The value is used later for conditional checks.

```python
service_type = input('Enter service level (Standard or Express): ')
print('Service level selected:', service_type)
```

#### Weight Classification
Uses if-elif-else to assign rates based on weight range.

```python
if weight <= 0:
    rate = 0.0
elif weight < 3.0:
    rate = 0.11 # Light rate
    print('Category: Standard Parcel')
elif weight < 5.0:
    rate = 0.15 # Medium rate
    print('Category: Medium Freight')
else:
    rate = 0.21 # Heavy rate
    print('Category: Heavy Cargo')
```

#### Nested Distance Surcharges
Uses nested if statements to apply additional fees based on distance. Surcharges vary by distance range.

**Calculate shipping subtotal using an expression**

```python
sub_shipping = weight * distance * rate
```

**Nested logic for distance surcharges**

```python
if distance > 100:
    print('Checking for regional surcharges...')
    if distance > 500:
        dist_surcharge = 45
        print('Long-haul surcharge applied.')
    else:
        dist_surcharge = 20
        print('Mid-range surcharge applied.')
else:
    dist_surcharge = 0.0
    print('No distance surcharge applied.')
```

#### Service Type Conditionals
Uses if-else to apply an additional fee based on the selected service type.

```python
if service_type == 'Express':
    added_service_cost = PRIORITY_FEE
    print('Priority processing added.')
else:
    added_service_cost = 0.0
    print('Standard processing selected.')
```

#### Final Cost Calculations
Uses arithmetic expressions to compute the total shipping cost.

**Total Cost**

```python
pre_tax = sub_shipping + dist_surcharge + added_service_cost + BASE_FEE
tax_amount = pre_tax * VAT_RATE
total_cost = pre_tax + tax_amount
```

**High-value orders**

```python
if total_cost > 500:
    print('Notice: This transaction requires manager approval.')
```

#### Invoice
Displays the final shipping summary and total cost for the user.

```python
if weight > 0 and distance > 0:
    print('---------- SHIPPING SUMMARY ----------')
    print('Calculated Rate:', rate)
    print('Shipping Subtotal:', sub_shipping)
    print('Distance Fee:', dist_surcharge)
    print('Service Fee:', added_service_cost)
    print('Tax Amount:', tax_amount)
    print('--------------------------------------')
    print('TOTAL PAYABLE:', total_cost)
    print('--------------------------------------')
else:
    print('Program Error: Could not compute total. Check input values.')
```
