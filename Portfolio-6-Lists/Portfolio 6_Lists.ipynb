## Food Preference Analytics System
<div style="text-align: justify;">
This program features a Food Preference and Analytics System which is an interactive program designed to compare local and international dishes through a structured voting process. The system uses lists to store food categories, loops to manage multiple matchups, and functions to organize the tournament flow. It applies conditionals for decision-making in vote counting, while user input allows multiple participants to evaluate each matchup using an A/B selection system. The program also incorporates randomization for handling invalid inputs and ensuring fair pairing of dishes. In addition, it computes statistical outputs such as total matches, win percentages, and most popular dishes, providing a simple form of data analysis based on user preferences.

#### Imports and Data
Defines the required Python module (random) and sets up the two lists of dishes: local and international.

```python
import random

local_dishes = ["Adobo", "Sinigang", "Lechon", "Sisig", "Chicken Inasal"]
international_dishes = ["Spaghetti", "Carbonara", "Pizza", "Fries", "Sushi"]
```

#### Players and Statistics Setup
Asks how many players will participate and initializes the statistics system for tracking wins, matches, and dish popularity.

```python
players = int(input("How many players will participate? "))

stats = {
    "local_wins": 0,
    "international_wins": 0,
    "matches": 0,
    "local_favorites": {},
    "international_favorites": {}
}

for dish in local_dishes:
    stats["local_favorites"][dish] = 0

for dish in international_dishes:
    stats["international_favorites"][dish] = 0
```

#### Match & Voting
Handles each match where players vote using A or B. It counts votes, determines the winner, and updates all statistics.

```python
def match(d1, d2):
    votes_d1 = 0
    votes_d2 = 0

    print(f"\nMatch: A) {d1} vs B) {d2}")

    for i in range(players):
        choice = input(f"Player {i+1} vote (A/B): ").strip().lower()

        if choice == "a":
            votes_d1 += 1
        elif choice == "b":
            votes_d2 += 1
        else:
            winner = random.choice([d1, d2])
            print(f"Invalid input from Player {i+1}, random vote assigned.")
            if winner == d1:
                votes_d1 += 1
            else:
                votes_d2 += 1

    if votes_d1 > votes_d2:
        winner = d1
    else:
        winner = d2

    stats["matches"] += 1

    if winner in local_dishes:
        stats["local_wins"] += 1
        stats["local_favorites"][winner] += 1
    else:
        stats["international_wins"] += 1
        stats["international_favorites"][winner] += 1

    print("Winner:", winner)
    return winner
```

#### Safe Round System
Controls how matches are created while ensuring no dish is matched against itself and no duplicate matchups occur in the same round.

```python
def play_round(list1, list2):
    winners = []

    random.shuffle(list1)
    random.shuffle(list2)

    used2 = set()

    for d1 in list1:

        valid_opponents = [d for d in list2 if d != d1 and d not in used2]

        if not valid_opponents:
            valid_opponents = [d for d in list2 if d != d1]

        d2 = random.choice(valid_opponents)
        used2.add(d2)

        winners.append(match(d1, d2))

    return winners
```

#### Run Tournaments
Runs the three tournament modes: local vs local, international vs international, and local vs international.

```python
print("LOCAL vs LOCAL")
play_round(local_dishes, random.sample(local_dishes, len(local_dishes)))

print("\nINTERNATIONAL vs INTERNATIONAL")
play_round(international_dishes, random.sample(international_dishes, len(international_dishes)))

print("\nLOCAL vs INTERNATIONAL")
play_round(local_dishes, international_dishes)
```

#### Results
Calculates and displays total matches, win percentages for local and international dishes, and overall performance.

```python
total = stats["matches"]

local_percent = (stats["local_wins"] / total) * 100 if total else 0
international_percent = (stats["international_wins"] / total) * 100 if total else 0

print("\nFINAL STATISTICS")
print("Players:", players)
print("Total matches:", total)
print("Local wins:", stats["local_wins"], f"({local_percent:.2f}%)")
print("International wins:", stats["international_wins"], f"({international_percent:.2f}%)")
```

#### Most Popular Dishes
Finds and displays which local and international dishes were most frequently chosen as winners.

```python
most_local = max(stats["local_favorites"], key=stats["local_favorites"].get)
most_international = max(stats["international_favorites"], key=stats["international_favorites"].get)

print("\nMOST POPULAR DISHES")
print("Most popular local dish:", most_local)
print("Most popular international dish:", most_international)
```

#### Overall Result
Compares total wins and declares whether local cuisine, international cuisine, or neither (tie) performed better overall.

```python
if stats["local_wins"] > stats["international_wins"]:
    print("\nOverall result: Local wins")
elif stats["international_wins"] > stats["local_wins"]:
    print("\nOverall result: International wins")
else:
    print("\nOverall result: Tie")
```
