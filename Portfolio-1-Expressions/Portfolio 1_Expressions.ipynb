## Description
This program calculates a student's allowance, expenses, and savings using expressions.

## Concepts Used
- Variables
- Expressions
- Operators (+, -, *, /)
- Type Conversion
- Input and Output

```python
# User's basic information

name = input("Enter your name: ")
days = input("Enter number of school days: ")
daily_allowance = input("Enter your daily allowance: ")

lunch = input("Enter daily lunch expense: ")
dinner = input("Enter daily dinner expense: ")
snacks = input("Enter daily snacks expense: ")
```

```python
# Inputs (strings) into numbers

days = float(days)
daily_allowance = float(daily_allowance)

lunch = float(lunch)
dinner = float(dinner)
snacks = float(snacks)
```

```python
# Calculations using expressions

total_allowance = days * daily_allowance

daily_expenses = lunch + dinner + snacks
total_expenses = daily_expenses * days

remaining_balance = total_allowance - total_expenses
average_spending = total_expenses / days

savings_percent = (remaining_balance / total_allowance) * 100
```

```python
# Results

print("\n--- Expense Summary ---")
print("Name:", name)
print("Total Allowance:", total_allowance)
print("Total Expenses:", total_expenses)
print("Remaining Balance:", remaining_balance)
print("Average Daily Spending:", average_spending)
print("Savings Percentage:", savings_percent, "%")
```
